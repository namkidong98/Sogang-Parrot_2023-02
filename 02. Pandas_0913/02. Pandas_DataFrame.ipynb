## 인덱싱
1. 특정 열 값을 추출: 인덱싱 또는 속성
 - 변수[열이름] : 지정된 열 추출
 - 변수[[열이름1,열이름2,...]]
 - 변수.속성
 
 
2. 리스트와 유사하게 슬라이싱으로 행 추출 
 - 변수[start:end] #start행부터 end-1행까지 추출
 
 
3. 특정 행을 추출
 - 변수.loc[행이름] #지정된 레이블의 행 추출
 - 변수.loc[ [행이름1,행이름2,..] ] #2개 이상은 리스트로 묶어 지정


4. 특정한 요소 하나만을 선택
 - 변수.loc[ 행이름, 열이름 ]


```python
# iloc[s:e , s:e] : 인덱스(번호)를 이용한 인덱싱
data.iloc[:,1:4] # row는 모두 포함, column은 1번부터 3번(4불포함)까지
```

```python
data.iloc[0:5,:]    # 0번째 행부터 4번째 행까지, column은 모두 포함
```

```python
# loc : 이름을 이용한 인덱싱

data.loc[['India','Malawi']] # 행의 index명을 가지고 인덱싱
```

## 행&열 추가 삭제 수정

```python
# 행 추가 : data.loc['index명'] = [각 column에 해당되는 값의 리스트]
# 열 추가 : data['column명'] = [각 row에 해당되는 값의 리스트]

data['pop']=[24,29,100,38,23,13,87,33] #열 추가 (수정도 동일하게)
data
```

```python
data=data.drop(['pop'], axis=1) #pop 열 삭제
data=data.drop('Uruguay') #우루과이 행 삭제
data
```

```python
# 행 추가 : data.loc['index명'] = [각 column에 해당되는 값의 리스트]
# 열 추가 : data['column명'] = [각 row에 해당되는 값의 리스트]

data.loc['Uruguay']=['URY',2000,3219.793,12.099592,2.525596e+04,78.978740,5.108068] #행 추가 (수정)
data
```

## DataFrame 결합
Series도 같은 방법으로 결합
- pd.merge(df1, df2, on='', how='') : 객체를 차례대로 적고, on은 기준 column, how는 결합 방식

- pd.concat([df1, df2], axis=) : 연결할 두 객체를 리스트로 묶어서 인자로 주어야함

```python
# merge
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value2': [5, 6, 7, 8]})

print(pd.merge(df1, df2, on='key', how='inner'), '\n')  # 'key' 열을 기준으로 내부 결합 (일치하는 것들만)
print(pd.merge(df1, df2, on='key', how='left'), '\n')   # 'key'열 기준으로 left 결합 (첫번째는 모두 찍고 두번째는 일치하는 것만, 빈 곳은 NaN을 삽입)
print(pd.merge(df1, df2, on='key', how='outer'))        # 'key'열 기준으로 모든 key를 살리되 빈 곳은 모두 NaN을 삽입
```

```python
#concat
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'B': ['B3', 'B4', 'B5'],
                    'D': ['D0', 'D1', 'D2']})
print(pd.concat([df1, df2], axis=0), '\n')   # 행 방향 기준
print(pd.concat([df1, df2], axis=1), '\n')   # 열 방향 기준
print(pd.concat((df1, df2), axis=0, join='inner'))  # inner = 일치하는 것만 기준으로, axis=0이라 행 방향으로 아래에 추가
```

## apply & map

```python
data = {'A': [1, 2, 3, 4],
        'B': [5, 6, 7, 8]}
df = pd.DataFrame(data)

print(df, '\n')
print(df.apply(sum), '\n')              # default로 axis=0이 적용, 행 끼리 sum을 하므로 열마다 값이 출력된다
print(df.apply(sum, axis=1), '\n')      # axis=1이므로 열 끼리 sum을 하게 된다 --> 행마다 값이 출력된다
print(df.apply(lambda x: x**2))         # 모든 값에 제곱 함수를 적용
```

```python
data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}
df = pd.DataFrame(data)
df['A'] = df['A'].map({1: 'One', 2: 'Two'}) # 'A' 열의 값을 대체
df      # 'A'열에 대해 1, 2만 mapping해서 남은 3에는 NaN이 적용됨
```

## replace & where
데이터 프레임을 결과로 반환하기 때문에 변화를 반영하려면 재할당 혹은 inplace 옵션을 사용해야 한다 
- replace : dictionary 형태로 값을 주어 기존 값을 대체

- where : 조건과 대체할 값 혹은 함수를 준다 --> 조건에 부합하지 '않는' 값들을 대체할 값 혹은 함수로 바꾼다

```python
df = pd.DataFrame({'A': [1, 2, 3, 4],
                   'B': [5, 6, 7, 8]})

df['A'].replace({1: 'One', 3: 'Three'}) # 'A' 열의 1을 'One', 3을 'Three'으로 대체, inplace로 변화 반영
```

```python
df.where(df['A'] > 2)        # 대체할 값 주어지지 않으면, False일 경우 -> Nan으로 자동 대체
```

```python
df.where(df['A'] > 2, 0)    # A열 값이 2보다 작은 경우 해당 값을 0으로 대체
```

```python
df.where(df['B']<7, df.apply(lambda x : -x)) # B열이 7보다 같거나 큰 경우, df값의 부호를 반대로
```
